```python
import pandas as pd
import numpy as np

ndvi_df = pd.read_csv("/mnt/d/SEM_six_SGP/Crop_Price_V2/NDVI/Gujarat_Monthly_NDVI_2005_2024.csv")
ndvi_df.head()
```

```python
print(ndvi_df.columns)
```

```python
ndvi_df = ndvi_df[[
    "ADM2_NAME",
    "mean",
    "year",
    "month"
]]
```

```python
ndvi_df = ndvi_df.rename(columns={
    "ADM2_NAME": "district",
    "mean": "monthly_ndvi_mean"
})
```

```python
ndvi_df["district"] = (
    ndvi_df["district"]
    .str.strip()
    .str.lower()
)
```

```python
print(ndvi_df["monthly_ndvi_mean"].describe())
```

```python
ndvi_df = ndvi_df[
    (ndvi_df["monthly_ndvi_mean"] >= -1) &
    (ndvi_df["monthly_ndvi_mean"] <= 1)
]
```

```python
ndvi_df["year"] = ndvi_df["year"].astype(int)
ndvi_df["month"] = ndvi_df["month"].astype(int)
```

```python
ndvi_df["year"] = ndvi_df["year"].astype(int)
ndvi_df["month"] = ndvi_df["month"].astype(int)
```

```python
print(ndvi_df.shape)
print(ndvi_df["district"].nunique())
print(ndvi_df.head())
```

```python
ndvi_df = ndvi_df.sort_values(
    by=["district", "year", "month"]
).reset_index(drop=True)

ndvi_df.to_csv(
    "/mnt/d/SEM_six_SGP/Crop_Price_V2/NDVI/gujarat_monthly_ndvi_clean_2005_2024_final.csv",
    index=False
)

print("Clean NDVI dataset saved successfully 🚀")
```

RAIN DATA

```python
rain_df = pd.read_csv("/mnt/d/SEM_six_SGP/Crop_Price_V2/rain/gujarat_rainfall_monthly_2005_2024.csv")

print(rain_df.shape)
print(rain_df.head())
```

```python
print(rain_df.columns)
```

```python
rain_df = rain_df[[
    "district",
    "year",
    "month",
    "monthly_rain_sum",
    "monthly_rain_mean"
]]
```

```python
rain_df["district"] = (
    rain_df["district"]
    .astype(str)
    .str.strip()
    .str.lower()
)
```

```python
rain_df["year"] = rain_df["year"].astype(int)
rain_df["month"] = rain_df["month"].astype(int)

rain_df["monthly_rain_sum"] = rain_df["monthly_rain_sum"].astype(float)
rain_df["monthly_rain_mean"] = rain_df["monthly_rain_mean"].astype(float)
```

```python
print(rain_df.isnull().sum())
```

```python
rain_df = rain_df[rain_df["monthly_rain_sum"] >= 0]
```

```python
rain_df = rain_df.sort_values(
    by=["district", "year", "month"]
).reset_index(drop=True)
```

```python
print(rain_df.shape)
print(rain_df["district"].nunique())
print(rain_df.head())
```

```python
rain_df.to_csv(
    "/mnt/d/SEM_six_SGP/Crop_Price_V2/rain/gujarat_monthly_rainfall_clean_2005_2024_final.csv",
    index=False
)

print("Clean rainfall dataset saved successfully 🚀")
```
